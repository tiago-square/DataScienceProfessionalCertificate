# Project: Generation of Geometric Shapes

This project, uses Python's matplotlib library to draw geometric shapes on screen. It demonstrates object-oriented programming concepts, showcasing the use of classes and objects.

```python
# Import matplotlib library
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class Circle(object):
    
    #Constructor
    def __init__(self,radius,color):
        self.radius = radius
        self.color = color

    #Method to add a radius value
    def addRadius(self,r):
        self.radius = self.radius + r
        return (self.radius)

    #Method to display the cicrle image
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radius, fc=self.color))
        plt.axis('scaled')
        plt.show() 
```

```python
# Create an object RedCircle
RedCircle = Circle(10,"Red")

# Call the method drawCircle
RedCircle.drawCircle()
```

```python
# Create an object BlueCircle
BlueCircle = Circle(3,"Blue")

# Call the method drawCircle
BlueCircle.drawCircle()
```

```python
# Create a new Rectangle class for creating a rectangle object

class Rectangle(object):
    
    # Constructor
    def __init__(self, width=2, height=3, color='r'):
        self.height = height 
        self.width = width
        self.color = color
    
    # Method
    def drawRectangle(self):
        plt.gca().add_patch(plt.Rectangle((0, 0), self.width, self.height ,fc=self.color))
        plt.axis('scaled')
        plt.show()
```

```python
# Create a new object rectangle
SkinnyBlueRectangle = Rectangle(2, 3, 'blue')

# Use the drawRectangle method to draw the shape
SkinnyBlueRectangle.drawRectangle()
```

```python
# Create a new object rectangle
FatYellowRectangle = Rectangle(20, 5, 'yellow')

# Use the drawRectangle method to draw the shape
FatYellowRectangle.drawRectangle()
```
